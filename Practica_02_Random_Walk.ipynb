Random Walk en 1D: se simula el precio de un activo

```python
# Importación de librerías
import random
import numpy as np
import matplotlib.pyplot as plt

# Nº periodos a simular
n_sim = 1000

# Precio inicial
precio_0 = 50

# Desviaciones
prob = 0.03

# Números aleatorios
rand_numbers = np.random.random(n_sim)

# Simulación de precios
precio = np.zeros(n_sim)
precio[0] = precio_0

for i in range(1, n_sim):
    if rand_numbers[i] < 0.5:
      precio[i] = precio[i-1] * (1+prob)
    else:
      precio[i] = precio[i-1] / (1+prob)

# Visualización gráfica
plt.title('Simulación del precio de un activo por Random Walk en 1D')
plt.plot(precio)
plt.show()
```

Random Walk en 2D

```python
# Importación de librerías
import random
import numpy as np
import matplotlib.pyplot as plt
import math

# Nº periodos a simular
n_sim = 1000

# Paso
paso = 1

# Números aleatorios
rand_numbers = np.random.random(n_sim)
angulo_r = rand_numbers*2*math.pi

# cálculo de "x" (coseno) e "y" (seno)
x = np.zeros(n_sim)
y = np.zeros(n_sim)

for i in range(1, n_sim):
  x[i] = x[i-1] + paso * math.cos(angulo_r[i])
  y[i] = y[i-1] + paso * math.sin(angulo_r[i])

# Visualización gráfica
plt.title('Random Walk en 2D')
plt.plot(x,y)
plt.show()
```
